```python
def load_data(directory):
    combined_data = pd.DataFrame()
    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            filepath = os.path.join(directory, filename)
            data = pd.read_csv(filepath)
            combined_data = pd.concat([combined_data, data], axis=0)
    return combined_data
```

```python
import os
import pandas as pd

# Define input and output directories
input_dir = "/Users/sahilkirti/mini-project(new)/Output"
output_dir = "/Users/sahilkirti/mini-project(new)/change_output"
```

```python
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
```

```python
csv_files = [f for f in os.listdir(input_dir) if f.endswith('.csv')]
```

```python
csv_files
```

```python
for csv_file in csv_files:
    # Read the CSV file skipping the header row
    df = pd.read_csv(os.path.join(input_dir, csv_file),header=None)
    header = df.iloc[0]
    df = df[1:]
    # Remove the last column
    df = df.iloc[:, :-1]
    # Remove '.csv' extension from filename and add '.txt'
    txt_filename = os.path.splitext(csv_file)[0] + '.txt'
    # Save the data to a text file in the output directory
    df.to_csv(os.path.join(output_dir, txt_filename), sep='\t', index=False , float_format='%.6f')
```
